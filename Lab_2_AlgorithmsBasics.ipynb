```python
# Exercise 5
def func5(n):
    # This function finds the integer square root of a given number 'n'.
    # It returns the integer 'x' such that x*x == n, or None if 'n' is not a perfect square.
    guess = 1
    while guess * guess <= n:
        if guess * guess == n:
            return guess
        guess = guess + 1
    return None
# Time Complexity: O(sqrt(N)), where N is the input number. The loop runs up to sqrt(N) times.
```

```python
# Sample function call: func5(16)
print(func5(16))
```

```python
# Exercise 7
def func12(arr, k):
    # This function rotates an array 'arr' to the right by 'k' positions.
    # Elements shifted out from the right end reappear at the left end.
    arr_len = len(arr)
    k = k % arr_len # Handle cases where k is greater than array length
    # Initialize a new array of the same length as 'arr'
    custom_arr = [0] * arr_len

    for i in range(arr_len):
        # Calculate the new position for each element after rotation.
        mystery_index = (i + k) % arr_len
        custom_arr[mystery_index] = arr[i]
    return custom_arr
# Time Complexity: O(N), where N is the length of the array. It iterates through the array once.
```

```python
#Sample function call: func12([1, 2, 3, 4, 5], 2)
print(func12([1,2,3,4,5],2))
```

```python
# write a function named as sum_arr to find sum of elements in given array
def sum_arr(arr):
  sum=0
  for i in range(len(arr)):
    sum+=arr[i]
  return sum
```

```python
# write a function named as Counting_EvenOdd to count number of even and odd elements in an array
def Counting_EvenOdd(arr):
  count_even=0
  count_odd=0
  for i in range(len(arr)):
    if arr[i]%2==0:
      count_even+=1
    else:
      count_odd+=1
      return count_even,count_odd
```

```python
# write a function named as linear_Search to perform a linear search in 1D-array
def linear_Search(arr,x):
  for i in range(len(arr)):
    if arr[i]==x:
      return i
  return -1
```
